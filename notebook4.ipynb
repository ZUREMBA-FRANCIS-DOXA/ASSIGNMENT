```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        print(f"Name: {self.name}, Age: {self.age}")


class Student(Person):
    def __init__(self, name, age, student_id):
        super().__init__(name, age)
        self.student_id = student_id

    def enroll_course(self, course_name):
        print(f"{self.name} (ID: {self.student_id}) has enrolled in {course_name}.")

    # Overridden display_info() — includes student_id, still uses parent's info
    def display_info(self):
        super().display_info()
        print(f"Student ID: {self.student_id}")


class Lecturer(Person):
    def __init__(self, name, age, employee_id):
        super().__init__(name, age)
        self.employee_id = employee_id

    def teach_course(self, course_name):
        print(f"{self.name} (Employee ID: {self.employee_id}) is teaching {course_name}.")


# --- Demonstration ---
student1 = Student("ZUREMBA FRANCIS DOXA", 20, "FOE.21.006.24.038")
lecturer1 = Lecturer("PROF. BERNARD KUMI BOATENG", 45, "UMaT/EMP/0123")

print("---- Student Info ----")
student1.display_info()      # Overridden version (includes student_id)
student1.enroll_course("OBJECT ORIENTED PROGRAMMING")

print("\n---- Lecturer Info ----")
lecturer1.display_info()     # Inherited, unmodified version from Person
lecturer1.teach_course("GEOLOGICAL ENGINEERING")
```
